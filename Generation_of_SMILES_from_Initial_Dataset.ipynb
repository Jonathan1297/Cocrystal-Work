```python
import pandas as pd
import requests

# Function to convert SMILES to IUPAC name using PubChem API
def smiles_to_iupac(smiles):
    url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/smiles/{smiles}/property/IUPACName/JSON"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        iupac_name = data['PropertyTable']['Properties'][0]['IUPACName']
        return iupac_name
    else:
        return None

# Create a pandas DataFrame with SMILES strings
df = pd.read_excel('Complete_Set_aug12.xlsx')

# Apply the smiles_to_iupac function to each SMILES string in the DataFrame
df['API_Name'] = df['API_SMILES'].apply(smiles_to_iupac)
df['Coformer_Name'] = df['Coformer_SMILES'].apply(smiles_to_iupac)


# Display the DataFrame with IUPAC names
df = df[['API_SMILES', 'API_Name','API_Molar Mass', 'Coformer_SMILES', 'Coformer_Name','Coformer_Molar Mass', 'Formation','Mass for 1:1 mmol']]
df.head()
```

```python
%%capture
!pip install mordred
!pip install rdkit-pypi
import pandas as pd
from rdkit import Chem

# Function to convert SMILES to Canonical SMILES
def convert_to_canonical_smiles(smiles_string):
    molecule = Chem.MolFromSmiles(smiles_string)
    if molecule is not None:
        canonical_smiles = Chem.MolToSmiles(molecule, isomericSmiles=True)
        return canonical_smiles
    else:
        return None

# Read the Excel file
file_name = 'smilesforall.xlsx'
df = pd.read_excel('smilesforall.xlsx')

# Assume your SMILES column is named 'SMILES' in the Excel file
smiles_column = 'Coformer_SMILES'

# Apply the conversion function to the SMILES column
df['Canonical_SMILES'] = df[smiles_column].apply(convert_to_canonical_smiles)

# Display the DataFrame with original and canonical SMILES
print(df[['Coformer_SMILES', 'coformer_Canonical_SMILES']])
```

```python
import pandas as pd
import requests

def get_canonical_smiles(molecule_name):
    base_url = "https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/"
    request_url = f"{base_url}{molecule_name}/property/CanonicalSMILES/JSON"

    response = requests.get(request_url)

    if response.status_code == 200:
        data = response.json()
        if 'PropertyTable' in data and 'Properties' in data['PropertyTable']:
            canonical_smiles = data['PropertyTable']['Properties'][0]['CanonicalSMILES']
            return canonical_smiles
        else:
            return f"Canonical SMILES data not available for {molecule_name}"
    else:
        return f"Error retrieving data for {molecule_name}. Status code: {response.status_code}"

# Read Excel file
df = pd.read_excel('New_Potential_API_Coformer_Pairs.xlsx')

# Add a new column for Canonical SMILES
df['API_SMILES'] = df['API_Name'].apply(get_canonical_smiles)
df['Coformer_SMILES'] = df['Coformer_Name'].apply(get_canonical_smiles)
df
```

```python
df1 = df.drop(['API_Name', 'Coformer_Name' ], axis=1)
df1
```

```python
df1.to_excel('New_Potential_API_COF_Pair_SMILES.xlsx', index=None)
```
