```python
def predict_form(smiles_API, smiles_COF):
    mol_list_API = []
    mol_list_COF = []

    # Convert SMILES strings to lists
    smiles_API = [smiles_API]
    smiles_COF = [smiles_COF]

    # Process SMILES_API
    for smile in smiles_API:
        mol = Chem.MolFromSmiles(smile)
        mol = Chem.AddHs(mol)
        AllChem.EmbedMolecule(mol)
        mol_list_API.append(mol)

    # Process SMILES_COF
    for smile in smiles_COF:
        mol = Chem.MolFromSmiles(smile)
        mol = Chem.AddHs(mol)
        AllChem.EmbedMolecule(mol)
        mol_list_COF.append(mol)

    # Create DataFrames for API and COF molecules
    df_mol_API = pd.DataFrame(data=mol_list_API, columns=['mol_API'], dtype='object')
    df_mol_COF = pd.DataFrame(data=mol_list_COF, columns=['mol_COF'], dtype='object')

    # Calculate descriptors for API and COF molecules
    calc = Calculator(descriptors, ignore_3D=False)
    desc_API = calc.pandas(df_mol_API['mol_API'])
    desc_COF = calc.pandas(df_mol_COF['mol_COF'])

    # Select only the specified descriptors for desc_API
    desc_API_selected = desc_API[['ATSC6i', 'PEOE_VSA8', 'VR1_A', 'ATSC3d',
                                  'SpAbs_A', 'AATS0dv', 'ATSC2d', 'PNSA4', 'EState_VSA7',
                                  'PNSA3', 'RPCS', 'SMR_VSA6', 'VSA_EState5', 'ATSC8dv',
                                  'ATSC4Z', 'PEOE_VSA2', 'ATSC3dv', 'PEOE_VSA7', 'EState_VSA3',
                                  'ATSC1v', 'ATSC8i', 'EState_VSA4', 'ATSC8v', 'SlogP_VSA3',
                                  'ATSC3s', 'ATSC1dv', 'RNCS', 'AATS0i', 'PPSA3', 'VSA_EState2',
                                  'ATSC5dv', 'ATSC4i', 'ATSC8Z', 'BCUTdv-1h', 'ATSC4v', 'ATSC5Z',
                                  'ATSC5d', 'ATSC8d', 'AATS0Z', 'ATSC1pe', 'AATSC4Z']]

    # Select only the specified descriptors for desc_COF
    desc_COF_selected = desc_COF[['AATS0dv', 'ATSC1v', 'AATS0Z', 'C1SP2', 'BCUTdv-1h', 'VSA_EState2',
                                  'BCUTdv-1l', 'AATS1Z', 'BCUTd-1h', 'IC1', 'AATSC0dv', 'SpDiam_A',
                                  'RNCS', 'AATSC1dv', 'ATSC2Z', 'SpAbs_A', 'AATSC2Z', 'RPCS', 'ATSC1i',
                                  'ATSC1dv', 'ATSC1p', 'ATSC2s', 'BCUTs-1h', 'AATSC2dv', 'AATSC0i',
                                  'DPSA1', 'nAromAtom', 'AATSC2s', 'BCUTv-1l']]

    desc_concat = pd.concat([desc_API, desc_COF], axis=1)


    # Return the selected descriptors for API and COF
    return desc_API_selected, desc_COF_selected

# Call the function with SMILES strings as strings
desc_API, desc_COF = predict_form('COC1=CC(=CC(=C1OC)OC)CC2=CN=C(N=C2N)N	', 'CC1=CC(=NC(=N1)NS(=O)(=O)C2=CC=C(C=C2)N)C	')

desc_COF.columns = [name+'_COF' for name in desc_COF.columns]

# concatenate the descriptors of apis and coformers and add target as well
df_finall = pd.concat([desc_API, desc_COF], axis=1)

# Reorder columns of df2 to match the order of df1
df_final = df_finall[X.columns]


```
